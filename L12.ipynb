```python
import pandas as pd
from pandas_datareader.data import get_data_yahoo
```

We're going to get price data of the following tickers.

IVV : S&P 500 ETF  
IDEV : International Developed Market ETF  
IUSB : Total USD Bond Market ETF  
IEMG : Emerging Markets ETF  
IAGG : International Aggregate Bond ETF  
IJH : S&P Mid-Cap ETF  
IJR : S&P Small-Cap ETF  

```python
data = get_data_yahoo(["IVV","IDEV","IUSB","IEMG","IAGG","IJH","IJR"], '2015-01-01', '2019-12-31')
data
```

```python
adj_close = data['Adj Close']
adj_close
```

```python
adj_close_returns = adj_close.pct_change().add_suffix("_return")
adj_close_returns
```

```python
adj_close.to_csv("AdClosePrices.csv")
adj_close_returns.to_csv('AdCloseReturn.csv')
```
